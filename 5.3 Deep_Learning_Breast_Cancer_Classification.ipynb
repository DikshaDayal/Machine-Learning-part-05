```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.datasets
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python


print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
X = data_frame.drop(columns='label' , axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state=2)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

image.png

```python
#import tenserflow and Keras
import tensorflow as tf
tf.random.set_seed(3)  #This one is for reproducibility
#By setting a random seed, you can ensure that your experiments are reproducible. This means that every time you run your code, you will get the same results, which is crucial for debugging and comparing results.
from tensorflow import keras
```

```python
#setting up the layer for Neural Network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),  # 30 features are there
                          keras.layers.Dense(20, activation='relu'),   #20 hidden preceptron
                          keras.layers.Dense(2, activation='sigmoid')   # 2 will represents the number of outputs , in our case it's either 0 or 1
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

sparse_categorical_crossentropy is used when the data is 0,1,2,3,......... form

categorical_crossentropy is used when the data is [1,0,0],[0,1,0],[0,0,1] form

```python
#trainning the neural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python

loss, accuracy = model.evaluate(X_test_std, Y_test)
print((round(accuracy*100)))
```

```python
#Find the maximum of the element
# argmax function

my_list = [0.25, 0.56]

index_of_max_value = np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

Building the predictive system

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
